```python
# Initialize lists to store the extracted data
placements = []
prizes = []
prize_percentages = []
team_names = []
rosters = []

# Loop through each row (skip the header row)
for row in rows[1:]:
    # Extract the placement
    placement_td = row.find('td', class_='tournament-results-place')
    if placement_td:
        placement = placement_td.text.strip()
    else:
        placement = 'N/A'  # Handle missing placement
    placements.append(placement)

    # Extract the prize
    prize_td = row.find('td', class_='tournament-results-prize')
    if prize_td:
        prize = prize_td.text.strip()
    else:
        prize = 'N/A'  # Handle missing prize
    prizes.append(prize)

    # Extract the prize percentage
    prize_percentage_td = row.find_all('td')[1]
    if prize_percentage_td:
        prize_percentage = prize_percentage_td.text.strip()
    else:
        prize_percentage = 'N/A'  # Handle missing prize percentage
    prize_percentages.append(prize_percentage)

    # Extract the team name
    team_name_span = row.find('span', class_='teamname')
    if team_name_span:
        team_name = team_name_span.text.strip()
    else:
        team_name = 'N/A'  # Handle missing team name
    team_names.append(team_name)

    # Extract the roster
    roster_td = row.find('td', class_='tournament-results-roster')
    if roster_td:
        roster = roster_td.text.strip().replace(' • ', ', ')
    else:
        roster = 'N/A'  # Handle missing roster
    rosters.append(roster)
```

```python
# Base URL
base_url = "https://cod-esports.fandom.com/wiki/Call_of_Duty_League/2023_Season/Major_{}/Qualifiers"

# List to store all data
all_team_data = []

# Set to keep track of team names
seen_team_names = set()
```

```python
# Function to parse and sum counts
def parse_and_sum(count_string):
    counts = count_string.split(' - ')
    return int(counts[0]), int(counts[1])
```

```python
def combine_counts(original_count, new_count):
    orig_wins, orig_losses = parse_and_sum(original_count)
    new_wins, new_losses = parse_and_sum(new_count)
    total_wins = orig_wins + new_wins
    total_losses = orig_losses + new_losses
    return f"{total_wins} - {total_losses}"
```

```python
# Loop through each stage
for stage in range(1, 6):
    url = base_url.format(stage)
    response = requests.get(url)
    page_soup = soup(response.content, 'html.parser')
    
    # List of search parameters
    search_classes = [
        'teamhighlight teamhighlighter',
        'standings-up teamhighlight teamhighlighter',
        'standings-stay teamhighlight teamhighlighter'
    ]
    
    # Loop through each search class to find all teams
    for class_name in search_classes:
        target_elements = page_soup.find_all('tr', class_=class_name)
        
        for results in target_elements:
            try:
                team_name = results.find('span', class_='teamname').text.strip()
            except:
                team_name = 'n/a'
            
            try:
                series_count = results.find_all('td')[2].text.strip()
            except:
                series_count = 'n/a'
            
            try:
                map_count = results.find_all('td')[4].text.strip()
            except:
                map_count = 'n/a'
            
            if team_name != 'n/a':
                if stage == 1:
                    # Add team names and their data during the first stage
                    seen_team_names.add(team_name)
                    all_team_data.append([team_name, series_count, map_count])
                else:
                    if team_name in seen_team_names:
                        # Update series and map counts for existing teams
                        for team_data in all_team_data:
                            if team_data[0] == team_name:
                                team_data[1] = combine_counts(team_data[1], series_count)
                                team_data[2] = combine_counts(team_data[2], map_count)
                                break
```

```python
#Convert the list to a DataFrame
team_data_df = pd.DataFrame(all_team_data, columns=['Team Name', '(Stages/Online - Series Count)', '(Stages/Online - Map Count)'])
```

```python
# Function to parse and sum counts from a series of count strings
def parse_and_sum_counts(count_series):
    total_wins = total_losses = 0
    for count in count_series:
        wins, losses = parse_and_sum(count)
        total_wins += wins
        total_losses += losses
    return total_wins, total_losses
```

```python
# Group by team name and aggregate the counts
stages_df = team_data_df.groupby('Team Name').agg({
    '(Stages/Online - Series Count)': lambda x: f"{parse_and_sum_counts(x)[0]} - {parse_and_sum_counts(x)[1]}",
    '(Stages/Online - Map Count)': lambda x: f"{parse_and_sum_counts(x)[0]} - {parse_and_sum_counts(x)[1]}"
}).reset_index()

stages_df
```

```python
#Split 'Series Count' and 'Map Count' to get win and loss counts
series_win_loss_stages = stages_df['(Stages/Online - Series Count)'].str.split(' - ', expand=True).astype(int)
map_win_loss_stages = stages_df['(Stages/Online - Map Count)'].str.split(' - ', expand=True).astype(int)
```

```python
# Calculate win percentages and round them
stages_df['(Stages/Online - Series Win Percentage)'] = (series_win_loss_stages[0] / (series_win_loss_stages[0] + series_win_loss_stages[1]) * 100).round(0)
stages_df['(Stages/Online - Map Win Percentage)'] = (map_win_loss_stages[0] / (map_win_loss_stages[0] + map_win_loss_stages[1]) * 100).round(0)
```

```python
# Insert new columns next to existing ones
stages_df.insert(stages_df.columns.get_loc('(Stages/Online - Series Count)') + 1, '(Stages/Online - Series Win Percentage)', stages_df.pop('(Stages/Online - Series Win Percentage)'))
stages_df.insert(stages_df.columns.get_loc('(Stages/Online - Map Count)') + 1, '(Stages/Online - Map Win Percentage)', stages_df.pop('(Stages/Online - Map Win Percentage)'))
stages_df
#online qualifiers dones 
```

```python
# Loop through each row in the majors_bk dataframe
for index, row in majors_bk.iterrows():
    team1 = row['Team 1']
    team2 = row['Team 2']
    score = row['Score']
    series_count = row['Series Count']
    
    # Split the score into individual map counts
    team1_maps, team2_maps = map(int, score.split(' - '))
    
    # Determine the series counts
    if team1_maps > team2_maps:
        team1_series = "1 - 0"
        team2_series = "0 - 1"
    else:
        team1_series = "0 - 1"
        team2_series = "1 - 0"
    
    # Append data for team 1
    team_names.append(team1)
    map_counts.append(f"{team1_maps} - {team2_maps}")
    series_counts.append(team1_series)
    
    # Append data for team 2
    team_names.append(team2)
    map_counts.append(f"{team2_maps} - {team1_maps}")
    series_counts.append(team2_series)
```

```python
#Step 4: Convert all filled NaNs to integers (if they aren't already)
allmajors_df['Map Wins_df'] = allmajors_df['Map Wins_df'].astype(int)
allmajors_df['Map Losses_df'] = allmajors_df['Map Losses_df'].astype(int)
allmajors_df['Map Wins_df1'] = allmajors_df['Map Wins_df1'].astype(int)
allmajors_df['Map Losses_df1'] = allmajors_df['Map Losses_df1'].astype(int)

allmajors_df['Series Wins_df'] = allmajors_df['Series Wins_df'].astype(int)
allmajors_df['Series Losses_df'] = allmajors_df['Series Losses_df'].astype(int)
allmajors_df['Series Wins_df1'] = allmajors_df['Series Wins_df1'].astype(int)
allmajors_df['Series Losses_df1'] = allmajors_df['Series Losses_df1'].astype(int)
```

```python
# Step 5: Sum the corresponding columns
allmajors_df['Map Wins'] = allmajors_df['Map Wins_df'] + allmajors_df['Map Wins_df1']
allmajors_df['Map Losses'] = allmajors_df['Map Losses_df'] + allmajors_df['Map Losses_df1']
allmajors_df['Series Wins'] = allmajors_df['Series Wins_df'] + allmajors_df['Series Wins_df1']
allmajors_df['Series Losses'] = allmajors_df['Series Losses_df'] + allmajors_df['Series Losses_df1']
```

```python
# Step 6: Combine the wins and losses back into 'Map Count' and 'Series Count' columns
allmajors_df['Map Count'] = allmajors_df['Map Wins'].astype(str) + ' - ' + allmajors_df['Map Losses'].astype(str)
allmajors_df['Series Count'] = allmajors_df['Series Wins'].astype(str) + ' - ' + allmajors_df['Series Losses'].astype(str)
```
