```python
#table that contains all of the champs teams 
#rosters 
#placements 
#prizes 
#rosters 

champs_bracket = s_object.find('table',class_='wikitable2 tournament-results zebra-manual')
champs_bracket
```

```python
# Find all rows in the table
rows = champs_bracket.find_all('tr')
rows
```

```python
# Initialize lists to store the extracted data
placements = []
prizes = []
prize_percentages = []
team_names = []
rosters = []

# Loop through each row (skip the header row)
for row in rows[1:]:
    # Extract the placement
    placement_td = row.find('td', class_='tournament-results-place')
    if placement_td:
        placement = placement_td.text.strip()
    else:
        placement = 'N/A'  # Handle missing placement
    placements.append(placement)

    # Extract the prize
    prize_td = row.find('td', class_='tournament-results-prize')
    if prize_td:
        prize = prize_td.text.strip()
    else:
        prize = 'N/A'  # Handle missing prize
    prizes.append(prize)

    # Extract the prize percentage
    prize_percentage_td = row.find_all('td')[1]
    if prize_percentage_td:
        prize_percentage = prize_percentage_td.text.strip()
    else:
        prize_percentage = 'N/A'  # Handle missing prize percentage
    prize_percentages.append(prize_percentage)

    # Extract the team name
    team_name_span = row.find('span', class_='teamname')
    if team_name_span:
        team_name = team_name_span.text.strip()
    else:
        team_name = 'N/A'  # Handle missing team name
    team_names.append(team_name)

    # Extract the roster
    roster_td = row.find('td', class_='tournament-results-roster')
    if roster_td:
        roster = roster_td.text.strip().replace(' • ', ', ')
    else:
        roster = 'N/A'  # Handle missing roster
    rosters.append(roster)

```

```python
data = {
    'Placement': placements,
    'Prize': prizes,
    'Prize (%)': prize_percentages,
    'Team Name': team_names,
    'Roster': rosters
}
```

```python
#chamsp roster and placing are done 
#now need bracket
#map series wins
champs_b = s_object.find('table',class_='wikitable md-table')
champs_b
```

```python
#define the search classes 
search_classes = [
    'mdv-allweeks mdv-week1',
    'mdv-allweeks mdv-week2',
    'mdv-allweeks mdv-week3'  
]
```

```python
teams = []
scores = []
series_count = {}

# Loop through the search classes to extract data for each week
for search_class in search_classes:
    rows = champs_b.find_all('tr', class_=search_class + ' toggle-section-hidden')
    
    # Loop through each row to extract team names and scores
    for row in rows:
        # Extract the team names
        team1 = row.find_all('span', class_='teamname')[0].text.strip()
        team2 = row.find_all('span', class_='teamname')[1].text.strip()
        
        # Extract the score
        score = row.find_all('td')[2].text.strip()  # The score is in the 3rd column (index 2)
        
        teams.append((team1, team2))
        scores.append(score)
        
        # Update series count
        score_values = list(map(int, score.split(' - ')))
        if score_values[0] > score_values[1]:
            winning_team = team1
            losing_team = team2
        else:
            winning_team = team2
            losing_team = team1
        
        # Increment series count
        if winning_team not in series_count:
            series_count[winning_team] = 0
        series_count[winning_team] += 1
```

```python
# Combine the extracted data into a DataFrame
champs_bk = pd.DataFrame({
    'Team 1': [team[0] for team in teams],
    'Team 2': [team[1] for team in teams],
    'Score': scores,
    'Series Count': [series_count.get(team, 0) for team in [team[0] for team in teams]]
})
champs_bk
```

```python
# Initialize a dictionary to hold aggregated results
team_stats = {}
```

```python
# Process each row in the DataFrame
for index, row in champs_bk.iterrows():
    team1, team2 = row['Team 1'], row['Team 2']
    score = row['Score']
    series_count = 1  # Each row represents one series

    team1_maps, team2_maps = map(int, score.split(' - '))
    
    # Initialize teams in the dictionary if they don't exist
    if team1 not in team_stats:
        team_stats[team1] = {'Map Count': [0, 0], 'Series Count': [0, 0]}
    if team2 not in team_stats:
        team_stats[team2] = {'Map Count': [0, 0], 'Series Count': [0, 0]}
    
    # Update map counts
    team_stats[team1]['Map Count'][0] += team1_maps
    team_stats[team1]['Map Count'][1] += team2_maps
    team_stats[team2]['Map Count'][0] += team2_maps
    team_stats[team2]['Map Count'][1] += team1_maps
    
    # Update series counts
    if team1_maps > team2_maps:
        team_stats[team1]['Series Count'][0] += 1  # Wins
        team_stats[team2]['Series Count'][1] += 1  # Losses
    else:
        team_stats[team1]['Series Count'][1] += 1  # Losses
        team_stats[team2]['Series Count'][0] += 1  # Wins
```

```python
# Convert the dictionary to a DataFrame
champs_bk_df = pd.DataFrame({
    'Team Name': team_stats.keys(),
    'Map Count': [f"{v['Map Count'][0]} - {v['Map Count'][1]}" for v in team_stats.values()],
    'Series Count': [f"{v['Series Count'][0]} - {v['Series Count'][1]}" for v in team_stats.values()]
})

# Display the resulting DataFrame
champs_bk_df
```

Now need to work on all the map count and series counts throughout Qualifiers and Major 
- online results

```python
# Base URL
base_url = "https://cod-esports.fandom.com/wiki/Call_of_Duty_League/2022_Season/Major_{}/Qualifiers"

# List to store all data
all_team_data = []

# Set to keep track of team names
seen_team_names = set()
```

```python
# Function to parse and sum counts
def parse_and_sum(count_string):
    counts = count_string.split(' - ')
    return int(counts[0]), int(counts[1])
```

```python
def combine_counts(original_count, new_count):
    orig_wins, orig_losses = parse_and_sum(original_count)
    new_wins, new_losses = parse_and_sum(new_count)
    total_wins = orig_wins + new_wins
    total_losses = orig_losses + new_losses
    return f"{total_wins} - {total_losses}"
```

```python
# Loop through each stage
for stage in range(1, 5):
    url = base_url.format(stage)
    response = requests.get(url)
    page_soup = soup(response.content, 'html.parser')
    
    # List of search parameters
    search_classes = [
        'standings-up teamhighlight teamhighlighter',
        'standings-stay teamhighlight teamhighlighter'
    ]
    
    # Loop through each search class to find all teams
    for class_name in search_classes:
        target_elements = page_soup.find_all('tr', class_=class_name)
        
        for results in target_elements:
            try:
                team_name = results.find('span', class_='teamname').text.strip()
            except:
                team_name = 'n/a'
            
            try:
                series_count = results.find_all('td')[2].text.strip()
            except:
                series_count = 'n/a'
            
            try:
                map_count = results.find_all('td')[4].text.strip()
            except:
                map_count = 'n/a'
            
            if team_name != 'n/a':
                if stage == 1:
                    # Add team names and their data during the first stage
                    seen_team_names.add(team_name)
                    all_team_data.append([team_name, series_count, map_count])
                else:
                    if team_name in seen_team_names:
                        # Update series and map counts for existing teams
                        for team_data in all_team_data:
                            if team_data[0] == team_name:
                                team_data[1] = combine_counts(team_data[1], series_count)
                                team_data[2] = combine_counts(team_data[2], map_count)
                                break
```

```python
#Convert the list to a DataFrame
team_data_df = pd.DataFrame(all_team_data, columns=['Team Name', '(Stages/Online - Series Count)', '(Stages/Online - Map Count)'])
```

```python
# Function to parse and sum counts from a series of count strings
def parse_and_sum_counts(count_series):
    total_wins = total_losses = 0
    for count in count_series:
        wins, losses = parse_and_sum(count)
        total_wins += wins
        total_losses += losses
    return total_wins, total_losses
```

```python
# Group by team name and aggregate the counts
stages_df = team_data_df.groupby('Team Name').agg({
    '(Stages/Online - Series Count)': lambda x: f"{parse_and_sum_counts(x)[0]} - {parse_and_sum_counts(x)[1]}",
    '(Stages/Online - Map Count)': lambda x: f"{parse_and_sum_counts(x)[0]} - {parse_and_sum_counts(x)[1]}"
}).reset_index()

stages_df
```

```python
#Split 'Series Count' and 'Map Count' to get win and loss counts
series_win_loss_stages = stages_df['(Stages/Online - Series Count)'].str.split(' - ', expand=True).astype(int)
map_win_loss_stages = stages_df['(Stages/Online - Map Count)'].str.split(' - ', expand=True).astype(int)
```

```python
# Calculate win percentages and round them
stages_df['(Stages/Online - Series Win Percentage)'] = (series_win_loss_stages[0] / (series_win_loss_stages[0] + series_win_loss_stages[1]) * 100).round(0)
stages_df['(Stages/Online - Map Win Percentage)'] = (map_win_loss_stages[0] / (map_win_loss_stages[0] + map_win_loss_stages[1]) * 100).round(0)
```

```python
# Insert new columns next to existing ones
stages_df.insert(stages_df.columns.get_loc('(Stages/Online - Series Count)') + 1, '(Stages/Online - Series Win Percentage)', stages_df.pop('(Stages/Online - Series Win Percentage)'))
stages_df.insert(stages_df.columns.get_loc('(Stages/Online - Map Count)') + 1, '(Stages/Online - Map Win Percentage)', stages_df.pop('(Stages/Online - Map Win Percentage)'))
stages_df
#online qualifiers dones 
```

```python
# Define the URL template and search classes
url_template = "https://cod-esports.fandom.com/wiki/Call_of_Duty_League/2022_Season/Major_{}"
search_classes = [
    'mdv-allweeks mdv-week1',
    'mdv-allweeks mdv-week2',
    'mdv-allweeks mdv-week3'
]
```

```python
# Initialize lists to hold the results
teams = []
scores = []
series_count = {}

# Loop through each Major (1 through 4)
for major_number in range(1, 5):
    url = url_template.format(major_number)
    response = requests.get(url)
    s_object = soup(response.text, 'html.parser')
    
    # Loop through the search classes to extract data for each week
    for search_class in search_classes:
        rows = s_object.find_all('tr', class_=search_class + ' toggle-section-hidden')
        
        # Loop through each row to extract team names and scores
        for row in rows:
            # Extract the team names
            team1 = row.find_all('span', class_='teamname')[0].text.strip()
            team2 = row.find_all('span', class_='teamname')[1].text.strip()
            
            # Extract the score
            score = row.find_all('td')[2].text.strip()  # The score is in the 3rd column (index 2)
            
            teams.append((team1, team2))
            scores.append(score)
            
            # Update series count
            score_values = list(map(int, score.split(' - ')))
            if score_values[0] > score_values[1]:
                winning_team = team1
                losing_team = team2
            else:
                winning_team = team2
                losing_team = team1
            
            # Increment series count
            if winning_team not in series_count:
                series_count[winning_team] = 0
            series_count[winning_team] += 1
```

```python
# Combine the extracted data into a DataFrame
majors_bk = pd.DataFrame({
    'Team 1': [team[0] for team in teams],
    'Team 2': [team[1] for team in teams],
    'Score': scores,
    'Series Count': [series_count.get(team, 0) for team in [team[0] for team in teams]]
})
majors_bk
```

```python
# Create lists to hold the new data
team_names = []
map_counts = []
series_counts = []
```

```python
# Loop through each row in the majors_bk dataframe
for index, row in majors_bk.iterrows():
    team1 = row['Team 1']
    team2 = row['Team 2']
    score = row['Score']
    series_count = row['Series Count']
    
    # Split the score into individual map counts
    team1_maps, team2_maps = map(int, score.split(' - '))
    
    # Determine the series counts
    if team1_maps > team2_maps:
        team1_series = "1 - 0"
        team2_series = "0 - 1"
    else:
        team1_series = "0 - 1"
        team2_series = "1 - 0"
    
    # Append data for team 1
    team_names.append(team1)
    map_counts.append(f"{team1_maps} - {team2_maps}")
    series_counts.append(team1_series)
    
    # Append data for team 2
    team_names.append(team2)
    map_counts.append(f"{team2_maps} - {team1_maps}")
    series_counts.append(team2_series)
```

```python
# Create the new dataframe
transformed_df = pd.DataFrame({
    'Team Name': team_names,
    'Map Count': map_counts,
    'Series Count': series_counts
})

# Display the transformed dataframe
transformed_df
```

```python
# Split the Map Count and Series Count into separate columns
transformed_df[['Map Wins', 'Map Losses']] = transformed_df['Map Count'].str.split(' - ', expand=True).astype(int)
transformed_df[['Series Wins', 'Series Losses']] = transformed_df['Series Count'].str.split(' - ', expand=True).astype(int)
```

```python
# Group by Team Name and sum the wins and losses
aggregated_df = transformed_df.groupby('Team Name').sum()
```

```python
# Combine the summed wins and losses back into the original formats
aggregated_df['Map Count'] = aggregated_df['Map Wins'].astype(str) + ' - ' + aggregated_df['Map Losses'].astype(str)
aggregated_df['Series Count'] = aggregated_df['Series Wins'].astype(str) + ' - ' + aggregated_df['Series Losses'].astype(str)

# Select the necessary columns
aggregated_df = aggregated_df[['Map Count', 'Series Count']].reset_index()
```

```python
# Rename the columns in the aggregated_df
aggregated_df.rename(columns={
    'Map Count': 'Major/LAN - Map Counts',
    'Series Count': 'Major/LAN - Series Counts'
}, inplace=True)
```

```python
# Create a dictionary to map abbreviated team names to full names
team_name_mapping = {
    'ATL': 'Atlanta FaZe',
    'BOS': 'Boston Breach',
    'FLA': 'Florida Mutineers',
    'LAG': 'Los Angeles Guerrillas',
    'LAT': 'Los Angeles Thieves',
    'LDN': 'London Royal Ravens',
    'MIN': 'Minnesota RØKKR',
    'NY': 'New York Subliners',
    'PAR': 'Paris Legion',
    'SEA': 'Seattle Surge',
    'TEX': 'OpTic Texas',
    'TOR': 'Toronto Ultra'
}

# Replace the abbreviated team names with full names
aggregated_df['Team Name'] = aggregated_df['Team Name'].replace(team_name_mapping)
```

```python
# Split the map counts and series counts into separate columns
aggregated_df[['Map Wins', 'Map Losses']] = aggregated_df['Major/LAN - Map Counts'].str.split(' - ', expand=True).astype(int)
aggregated_df[['Series Wins', 'Series Losses']] = aggregated_df['Major/LAN - Series Counts'].str.split(' - ', expand=True).astype(int)

# Calculate the win percentages
aggregated_df['Major/LAN - Map Win Percentage'] = (aggregated_df['Map Wins'] / (aggregated_df['Map Wins'] + aggregated_df['Map Losses'])) * 100
aggregated_df['Major/LAN - Series Win Percentage'] = (aggregated_df['Series Wins'] / (aggregated_df['Series Wins'] + aggregated_df['Series Losses'])) * 100

# Format the percentages to two decimal places
aggregated_df['Major/LAN - Map Win Percentage'] = aggregated_df['Major/LAN - Map Win Percentage'].round(2)
aggregated_df['Major/LAN - Series Win Percentage'] = aggregated_df['Major/LAN - Series Win Percentage'].round(2)

# Drop the intermediate columns used for calculations
aggregated_df.drop(columns=['Map Wins', 'Map Losses', 'Series Wins', 'Series Losses'], inplace=True)

# Reorder columns to place the new columns next to their respective counts
aggregated_df = aggregated_df[['Team Name', 'Major/LAN - Map Counts', 'Major/LAN - Map Win Percentage', 'Major/LAN - Series Counts', 'Major/LAN - Series Win Percentage']]
```
